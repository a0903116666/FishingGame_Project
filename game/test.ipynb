```python
import random
import pygame
import sys
from config import *

pygame.init()

screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("2D 體感釣魚大師 - 全螢幕模式")

clock = pygame.time.Clock()

fish_image = pygame.image.load(str("pixel_fish.png")).convert_alpha()

running = True

total_score = 0
font = pygame.font.SysFont(None, 36)

progress = 0
box_y = 300
box_speed = 100
box_size = 100
cooldown = 2
progress_corlor = (255, 0, 0)

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_ESCAPE:
                running = False

    screen.fill((0, 105, 255))

    _, cursor_y = body_cursor()
    cursor_pixel = max(min((cursor_y + 0.5) * WINDOW_HEIGHT, WINDOW_HEIGHT - 50), 50)
    
    # check if cursor is in the box area
    if abs(cursor_pixel - box_y) < box_size // 2:
        progress += 0.2 * 1 / FPS
        progress_corlor = (0, 255, 0)
    else:
        progress -= 0.4 * 1 / FPS
        progress_corlor = (255, 0, 0)
    
    progress = max(0, min(1, progress))

    if cooldown > 0:
        cooldown -= 1 / FPS
    if box_y >= WINDOW_HEIGHT - 50 - box_size // 2:
        box_y = WINDOW_HEIGHT - 49 - box_size // 2
        box_speed = -random.uniform(150, 300)
        cooldown = random.uniform(0.2, 2)
    elif box_y <= 50 + box_size // 2:
        box_y = 51 + box_size // 2
        box_speed = random.uniform(150, 300)
        cooldown = random.uniform(0.2, 2)
    else:
        if cooldown <= 0:
            box_speed = random.uniform(150, 300) * random.choice([-1, 1])
            cooldown = random.uniform(0.2, 2)
    box_y += box_speed * 1 / FPS


    # box render
    pygame.draw.rect(
        surface=screen,
        color=(150, 0, 150),
        rect=(50, box_y - box_size//2, 30, box_size),
        width=0
    )
    pygame.draw.rect(
        surface=screen, 
        color=(255, 165, 0), 
        rect=(50, 50, 30, WINDOW_HEIGHT - 100),
        width=2
    )

    # draw cursor with circle
    pygame.draw.circle(
        surface=screen, 
        color=(255, 0, 0), 
        center=(65, int(cursor_pixel)),
        radius=10
    )


    # progress bar render
    pygame.draw.rect(
        surface=screen,
        color=progress_corlor,
        rect=(WINDOW_WIDTH - 80, WINDOW_HEIGHT - (WINDOW_HEIGHT - 100) * progress - 50, 30, (WINDOW_HEIGHT - 100) * progress),
        width=0
    )
    pygame.draw.rect(
        surface=screen, 
        color=(255, 165, 0), 
        rect=(WINDOW_WIDTH - 80, 50, 30, WINDOW_HEIGHT - 100),
        width=2
    )

    pygame.display.flip()
    clock.tick(FPS)
    if cooldown > 0:
        cooldown -= 1 / FPS


pygame.quit()
sys.exit()
```
